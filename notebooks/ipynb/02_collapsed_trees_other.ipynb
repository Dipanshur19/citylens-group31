```python
from ultralytics import YOLO

model = YOLO("runs/detect/citylens_v1-2/weights/best.pt")
metrics = model.val(data="Fallen-trees-(with-palms)-2/data.yaml", split="test")

print(f"mAP@0.5:   {metrics.box.map50:.3f}")
print(f"Precision: {metrics.box.p.mean():.3f}")
print(f"Recall:    {metrics.box.r.mean():.3f}")
import cv2, json, os
from ultralytics import YOLO

model = YOLO("runs/detect/citylens_v1-2/weights/best.pt")

input_dir  = "Fallen-trees-(with-palms)-2/test/images"
output_dir = "output_frames"
os.makedirs(output_dir, exist_ok=True)

all_results = []

for img_name in os.listdir(input_dir):
    img_path = os.path.join(input_dir, img_name)
    results  = model(img_path)[0]

    annotated = results.plot()
    cv2.imwrite(os.path.join(output_dir, img_name), annotated)

    boxes_data = []
    for box in results.boxes:
        x, y, w, h = box.xywh[0].tolist()
        boxes_data.append({
            "class":      results.names[int(box.cls)],
            "confidence": round(float(box.conf), 3),
            "x": round(x,2), "y": round(y,2),
            "width": round(w,2), "height": round(h,2)
        })
    all_results.append({"image": img_name, "detections": boxes_data})

with open("bounding_boxes.json", "w") as f:
    json.dump(all_results, f, indent=2)

print("Done!")
```
